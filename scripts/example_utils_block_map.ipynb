```python
import numpy as np
import leopardgecko.utils as lgu
import matplotlib.pyplot as plt
```

```python
# define 3d array

a = np.zeros((20,20,20))
```

```python
# define a function to map
f_count=0
def vol_map_f(vol):
    global f_count
    b=np.ones_like(vol)*f_count

    #Put a marker on the centre
    c = [i//2 for i in b.shape]

    b[ c[0]-1:c[0]+1, c[1]-1:c[1]+1, c[2]-1:c[2]+1 ] = 0

    f_count+=1
    return b
```

```python
# map
f_count=0
c = lgu.map_vol_function_by_blocking( vol_map_f, a, (8,8,8), (2,2,2) )
```
