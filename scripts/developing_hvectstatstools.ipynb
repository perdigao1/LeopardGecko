```python
import numpy as np
import leopardgecko as lg
```

```python
hvectors = lg.hvectstatstools.get_hvect_combinations(2,12)
hvectors
```

```python
# Create numpy array with numembers from 0 to 12, being the count number from 2-class 12-way predictions

data= np.random.randint(0,13,size=(256,256,256))
```

```python
# create a ground truth binary-like data
gndtruth = np.random.randint(0,2,size=(256,256,256))
```

```python
hvect_gndclass_counter_in_data_counted = lg.hvectstatstools.gethvect_gndclass_counter_in_data_counted(data,gndtruth, hvectors)
```

```python
def gethvect_gndclass_counter_in_data_counted1(data, data_gnd, hvectors):
    print(data.size)

    nclasses = len(hvectors[0])
    nways = int(np.sum(np.array(hvectors[0],dtype=np.uint8)))
    print(f"nclasses:{nclasses} , nways:{nways}")
    #Initialise the counter
    hvect_gndclass_counter0 = {}
    for h0 in hvectors:
        d0= []
        for c0 in range(nclasses):
            d0.append(0)
        hvect_gndclass_counter0[ h0 ] = d0

    data_gnd_binary = (data_gnd!=0)

    #iterate calculation per vector basis
    for i0 in range(nways+1):
        hv0 = (i0, nways-i0)

        #get matches
        d_match = (data==nways-i0)
        match_count = np.count_nonzero(d_match)

        #Compare with gntruth
        count_compare_to_gnd = np.count_nonzero(np.logical_and(d_match,data_gnd_binary))
        print(f"i0:{i0} , match_count:{match_count}, count_compare_to_gnd:{count_compare_to_gnd}")

        hvect_gndclass_counter0[hv0] = [ match_count- count_compare_to_gnd, count_compare_to_gnd]

    return hvect_gndclass_counter0
```

```python
#Try ne method, hopefully faster
hvect_gndclass_counter_in_data_counted1 = gethvect_gndclass_counter_in_data_counted1(data,gndtruth, hvectors)
```
